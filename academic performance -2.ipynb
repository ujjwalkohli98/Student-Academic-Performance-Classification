```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('xAPI-Edu-Data.csv')
data.head()
```

```python
data['gender'].replace('M', 0,inplace=True)
data['gender'].replace('F', 1,inplace=True)

# Or we can use get_dummies to convert categorical values and concatenate them later
nat = pd.get_dummies(data['NationalITy'])
sid = pd.get_dummies(data['StageID'])
gid = pd.get_dummies(data['GradeID'])
secid = pd.get_dummies(data['SectionID'])
topic = pd.get_dummies(data['Topic'])
semester = pd.get_dummies(data['Semester'])
rel = pd.get_dummies(data['Relation'])
pas = pd.get_dummies(data['ParentAnsweringSurvey'])
pss = pd.get_dummies(data['ParentschoolSatisfaction'])
sab = pd.get_dummies(data['StudentAbsenceDays'])
```

```python
# Concatenating the variables we created above 
df = pd.concat([data,nat,sid,gid,secid,topic,semester,rel,pas,pss,sab],axis=1)
```

```python
df.drop(['NationalITy','PlaceofBirth','PlaceofBirth','StageID','GradeID','SectionID','Topic','Semester',
              'Relation','ParentAnsweringSurvey','ParentschoolSatisfaction','StudentAbsenceDays'],axis=1,inplace=True)
```

```python
df.columns
```

```python
Label = df['Class'] # Class is the value we want to predict

Features = df[['gender', 'raisedhands', 'VisITedResources', 'AnnouncementsView',
       'Discussion','Father', 'Mum', 'No', 'Yes', 'Bad', 'Good',
       'Above-7', 'Under-7']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(Features)
scaled = scaler.transform(Features)
```

```python
pd.DataFrame(scaled,columns=Features.columns)
```

```python
from sklearn.model_selection import train_test_split
X = scaled
y = Label
# split the data to 20% test,80% train with random state=42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
rfc = RandomForestClassifier(n_estimators=80,max_features='auto', max_depth=9,min_samples_leaf=1,
                             min_samples_split=2,bootstrap=True, random_state = 42)
rfc.fit(X_train,y_train)
rfcpred = rfc.predict(X_test)
print(classification_report(y_test,rfcpred))
```
